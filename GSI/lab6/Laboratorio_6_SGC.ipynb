## Objetivo

Implementar um pequeno SGC em Python cobrindo tres etapas do ciclo de vida do conhecimento:

- Producao: converter conhecimento tacito em explicito.
- Armazenamento: registrar o conhecimento em uma base estruturada.
- Disseminacao e aplicacao: recuperar informacao e usar regras para decisao automatica.


## Contexto do problema

O especialista Sr. Carlos explicou como resolver a lentidao do ERP nas segundas-feiras. O papel do SGC e evitar a amnesia corporativa, transformando essa experiencia em conhecimento acessivel para a organizacao.


```python
from pprint import pprint

from sgc_empresa import (
    buscar_conhecimento,
    converter_tacito_para_explicito,
    diagnostico_servidor,
    montar_base_exemplo,
    salvar_base_json,
)
```

# Parte 1 - Producao e Armazenamento

Primeiro, o relato verbal do especialista e convertido em um POP. Esse e o momento em que o conhecimento tacito passa a ser explicito.


```python
pop = converter_tacito_para_explicito()
pprint(pop, sort_dicts=False)
```

Em seguida, esse conhecimento e armazenado em uma lista de dicionarios. Tambem adicionamos novos registros, incluindo a falha fisica do rack e a regra de monitoramento de temperatura.


```python
base = montar_base_exemplo()
pprint(base, sort_dicts=False)
```

```python
arquivo = salvar_base_json(base)
print(f"Base salva em: {arquivo}")
```

## Parte 2 - Disseminacao

O conhecimento precisa ser encontrado rapidamente. Aqui usamos `list comprehension` para filtrar a base por palavra-chave.

A expressao tem a forma:

`[item for item in base if condicao]`

No laboratorio, a condicao verifica se o termo aparece no titulo, na causa ou na solucao.


```python
for termo in ["backup", "cabo", "temperatura"]:
    print(f"\nBusca por: {termo}")
    pprint(buscar_conhecimento(base, termo), sort_dicts=False)
```

## Parte 3 - Aplicacao

Agora o conhecimento deixa de ser apenas consulta e passa a orientar acao automatica. O motor de regras classifica o status do servidor com base na temperatura.


```python
for temperatura in [34.2, 35.7, 37.4]:
    print(f"{temperatura:.1f}C -> {diagnostico_servidor(temperatura)}")
```
