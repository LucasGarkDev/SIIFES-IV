## Objetivo
Implementar um protótipo de SGC utilizando Arquitetura em Camadas:

- Camada de Dados (Persistência em JSON)
- Camada de Lógica (Busca inteligente)
- Camada de Apresentação (Dashboard)

---

## Carregamento dos Dados

Nesta etapa, carregamos a base de conhecimento armazenada em JSON.
Caso o arquivo não exista, o sistema cria uma base inicial.

```python
from dados import carregar_dados

base = carregar_dados()

display("Quantidade de registros:", len(base))
base
```

## Busca de Conhecimento (Camada de Lógica)

A função implementa:
- Busca por título e conteúdo
- Case-insensitive
- Uso de List Comprehension

```python
from logica import buscar_conhecimento
resultados = buscar_conhecimento(base, "vpn")

for r in resultados:
    display(f"ID: {r['id']}")
    display(f"Título: {r['titulo']}")
    display(f"Conteúdo: {r['conteudo']}")
    display("-" * 40)
```

## Dashboard de Popularidade

Representação visual dos acessos usando barras.
Isso permite identificar rapidamente os conhecimentos mais utilizados.

```python
from apresentacao import exibir_dashboard
exibir_dashboard(base)
```

## Inserção de Novo Conhecimento

Simulando a adição de um novo item na base.

```python
from dados import salvar_dados
novo_item = {
    "id": max([item['id'] for item in base]) + 1,
    "titulo": "Erro de Login",
    "conteudo": "Resetar senha via sistema interno.",
    "acessos": 0
}

base.append(novo_item)
salvar_dados(base)

display("Novo conhecimento adicionado!")
```

## Validação da Persistência

Recarregando os dados para verificar se foram salvos corretamente.

```python
base_atualizada = carregar_dados()
base_atualizada[-1]
```

## Conclusão

O sistema implementado demonstra:

- Separação clara entre camadas
- Persistência de dados em JSON
- Recuperação eficiente de conhecimento

O protótipo atende aos requisitos de um Sistema de Gestão do Conhecimento básico.
